```python

# Step 2: Upload and extract dataset
from google.colab import files
uploaded = files.upload()  # Upload 'employee.zip'

import zipfile, os

with zipfile.ZipFile("employee.zip", 'r') as zip_ref:
    zip_ref.extractall("/content/")

```

```python
# Step 3: Preprocess data
import cv2
import numpy as np
from tensorflow.keras.utils import to_categorical

dataset_path = "/content/employee"
categories = [ "sleeping_image", "using_phone", "working_images"]

data = []
labels = []

for i, category in enumerate(categories):
    path = os.path.join(dataset_path, category)
    for img_name in os.listdir(path):
        try:
            img_path = os.path.join(path, img_name)
            img = cv2.imread(img_path)
            img = cv2.resize(img, (64, 64))
            data.append(img)
            labels.append(i)
        except:
            pass

data = np.array(data) / 255.0
labels = to_categorical(labels)
```

```python
# Step 4: Train model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))
```

```python
# ✅ Step 5: Image Prediction
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from PIL import Image

uploaded = files.upload()
filename = next(iter(uploaded))

img = image.load_img(filename, target_size=(64, 64))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

prediction = model.predict(img_array)
class_names = ['Employee Sleeping', 'Employee Using Phone', 'Employee Working']
predicted_class = class_names[np.argmax(prediction)]
confidence = np.max(prediction) * 100

print("Predicted:", predicted_class)
print("Confidence:", confidence)

original_img = Image.open(filename)
plt.imshow(original_img)
plt.title(f"{predicted_class} ({confidence:.2f}%)")
plt.axis('off')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Convert one-hot to class indices
y_true = np.argmax(y_test, axis=1)  # from one-hot to labels
y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

# Accuracy
accuracy = np.sum(y_pred == y_true) / len(y_true)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Classification Report
print("\nClassification Report:")
y_true = np.argmax(y_test, axis=1)


# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_true, y_pred))

# ROC-AUC Score (One-vs-Rest for multi-class)
try:
    roc_auc = roc_auc_score(y_test, y_pred_prob, multi_class='ovr')
    print(f"\nROC-AUC Score: {roc_auc:.4f}")
except Exception as e:
    print(f"\nROC-AUC calculation error: {e}")

# Plot ROC Curve for each class
y_true_bin = label_binarize(y_true, classes=range(len(class_names)))

plt.figure(figsize=(10, 8))
for i in range(len(class_names)):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Employee Behavior Class')
plt.legend(loc='lower right')
plt.show()
```

```python
# ✅ Step 6: Advanced Video Prediction (Multi-person detection + activity labels)

from IPython.display import display, Image as IPImage
import tempfile
from ultralytics import YOLO  # YOLOv8 import

# Load YOLO model (detect persons)
yolo_model = YOLO("yolov8n.pt")  # Nano model - fast and accurate

# Upload video
uploaded = files.upload()
video_path = next(iter(uploaded))

# Open video
cap = cv2.VideoCapture(video_path)

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = yolo_model(frame)  # detect objects
    boxes = results[0].boxes.xyxy.cpu().numpy()

    frame_copy = frame.copy()  # make copy to draw on

    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = [int(val) for val in box]
        person_img = frame[y1:y2, x1:x2]  # crop person

        try:
            resized = cv2.resize(person_img, (64, 64)) / 255.0
            reshaped = np.expand_dims(resized, axis=0)
            prediction = model.predict(reshaped)
            index = np.argmax(prediction)
            label = f"{class_names[index]} ({prediction[0][index]*100:.1f}%)"

            # Draw box & label
            cv2.rectangle(frame_copy, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame_copy, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                        0.6, (255, 255, 0), 2)
        except:
            continue

    # Show in Colab
    is_success, buffer = cv2.imencode(".jpg", frame_copy)
    if is_success:
        with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as f:
            f.write(buffer)
            display(IPImage(filename=f.name))

    frame_count += 1
    if frame_count > 30:
        break

cap.release()
```
